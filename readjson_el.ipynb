```python
import json
import pandas as pd

# Load the JSON file
with open('/Users/elinazachariou/Desktop/UCY Data Science/Semester 2/Capstone Project/sneakers_sample_19062025.txt', 'r') as f:
    raw_data = json.load(f)

# Create separate dataframes
meta_rows = []
data_rows = []
validation_rows = []
sync_rows = []
sources_rows = []

for entry in raw_data:
    # Top-level fields
    meta = {
        key: entry[key] for key in entry
        if key not in ['data', 'validation', 'synchronization', 'sources']
    }
    meta_rows.append(meta)
    
    # Extract nested fields (if they exist)
    if 'data' in entry:
        data_rows.append(entry['data'])
    if 'validation' in entry:
        validation_rows.append(entry['validation'])
    if 'synchronization' in entry:
        sync_rows.append(entry['synchronization'])
    if 'sources' in entry and isinstance(entry['sources'], list):
        for source in entry['sources']:
            # Optionally, attach product ID to track back
            source['mirakl_product_id'] = entry.get('mirakl_product_id')
            sources_rows.append(source)

# Convert all parts to DataFrames
df_meta = pd.DataFrame(meta_rows)
df_data = pd.json_normalize(data_rows)
df_validation = pd.json_normalize(validation_rows)
df_sync = pd.json_normalize(sync_rows)
df_sources = pd.DataFrame(sources_rows)

# Optional preview
print("Meta columns:", df_meta.columns.tolist())
print("Data columns:", df_data.columns.tolist())
print("Validation columns:", df_validation.columns.tolist())
print("Sources sample:", df_sources.head())
```

```python
# Flatten 'selling_authorization' column
selling_auth_df = pd.json_normalize(df_meta['selling_authorization'])

# Flatten 'data_origin' column
data_origin_df = pd.json_normalize(df_meta['data_origin'])

# Optional: Add prefixes so you know where they came from
selling_auth_df = selling_auth_df.add_prefix('selling_auth_')
data_origin_df = data_origin_df.add_prefix('origin_')

# Drop original nested columns from df_meta
df_meta = df_meta.drop(columns=['selling_authorization', 'data_origin'])

# Merge the flattened parts back
df_meta = pd.concat([df_meta, selling_auth_df, data_origin_df], axis=1)
```
